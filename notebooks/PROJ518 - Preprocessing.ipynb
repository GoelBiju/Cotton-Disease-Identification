```python
import numpy as np
from skimage import io
import os
from PIL import Image

import glob
from os.path import basename

from keras.preprocessing.image import ImageDataGenerator

import splitfolders
```

```python
# Data source
from google.colab import drive

# Importing the dataset (stored in Google Drive)
drive.mount("/content/drive", force_remount=True)

!ls "/content/drive/MyDrive/PROJ518/Cotton Dataset/raw"
raw_dir = "/content/drive/MyDrive/PROJ518/Cotton Dataset/raw"
```

```python
# Rename all the files in both healthy and leaf_curl folders
folder_name = 'healthy'
image_files = glob.glob(os.path.join(raw_dir, folder_name, '*.*'))

for i in range(len(image_files)):                          
  image_name = image_files[i]
  split_name = basename(image_files[i]).split('.')
  n = split_name[0]
  ext = split_name[-1]
  new_name = folder_name + '_' + str(i + 1) + '.' + ext
  print(new_name)

  os.rename(image_name, os.path.join(raw_dir, folder_name, new_name))
```

```python
# Construct an instance of the ImageDataGenerator class
# Pass the augmentation parameters through the constructor. 

datagen = ImageDataGenerator(
      rotation_range=45,     # Random rotation between 0 and 45
      width_shift_range=0.2, # % shifts
      height_shift_range=0.2,
      shear_range=0.2,
      zoom_range=0.1,
      horizontal_flip=True,
      vertical_flip=True,
      brightness_range=[0.2, 1.0],
      fill_mode='reflect', # nearest, constant, reflect, wrap

      # featurewise_center=False,             # Set input mean to 0 over the dataset
      # samplewise_center=False,              # Set each sample mean to 0
      # featurewise_std_normalization=False,  # Divide each input by its std
      # zca_whitening=False,                  # Apply ZCA whitening
      # zca_epsilon=1e-06,                    # Epsilon for ZCA whitening
)
```

```python
image_directory = raw_dir + '/leaf_curl/'
SIZE = 256
dataset = []
```

```python
my_images = os.listdir(image_directory)
print(len(my_images))

for i, image_name in enumerate(my_images):
    if (image_name.split('.')[1] in ['jpg', 'png', 'jpeg']):
        image = io.imread(image_directory + image_name)
        image = Image.fromarray(image, 'RGB')
        image = image.resize((SIZE, SIZE))
        dataset.append(np.array(image))

x = np.array(dataset)
```

```python
print(len(dataset))
print(len(x))
```

```python
# Generate augmented images and place into a new directory
os.mkdir(image_directory + '/augmented-leaf_curl')

i = 0
for batch in datagen.flow(x, batch_size=16,  
                          save_to_dir=image_directory + '/augmented-leaf_curl', 
                          save_prefix='aug_lc'):
    i += 1
    if i == 52:
        break

print(len(os.listdir(image_directory + '/augmented-leaf_curl')))
```

```python
# Get total counts of all data
healthy = next(os.walk(raw_dir + '/healthy'))[2]
print("Healthy: ", len(healthy))

leaf_curl = next(os.walk(raw_dir + '/leaf_curl'))[2]
print("Leaf Curl: ", len(leaf_curl))

aug_leaf_curl = next(os.walk(raw_dir + '/leaf_curl/augmented-leaf_curl'))[2]
print("Augmented Leaf Curl: ", len(aug_leaf_curl))
```
